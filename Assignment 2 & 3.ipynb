```python
employee_attrition2=employee_attrition.sort_values(by='DistanceFromHome',ascending=True)
cols = ['Attrition','Department','DistanceFromHome']
employee_attrition6=employee_attrition2[cols]
employee_attrition6[(employee_attrition6.Attrition=='Yes')&(employee_attrition6.Department=='Research & Development')]

```

```python
# Show a breakdown of distance from home by job role and attrition.
employee_attrition2=employee_attrition.sort_values(by='DistanceFromHome',ascending=True)
cols = ['Attrition','Department','DistanceFromHome']
employee_attrition2[cols]
employee_attrition6=employee_attrition2[cols]
employee_attrition6[(employee_attrition6.Attrition=='Yes')]
employee_attrition6.groupby('Department').DistanceFromHome.agg(['count','mean','max','min'])
```

```python
#  Compare average monthly income by education and attrition. We need the Monthly income column, Education column and attrition column

employee_attrition.sort_values(by='MonthlyIncome',ascending=True)
employee_attrition3=employee_attrition.sort_values(by='MonthlyIncome',ascending=True)
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition3[cols]
employee_attrition4=employee_attrition3[employee_attrition3.Education==1]
employee_attrition4
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition4[cols]
employee_attrition4.sort_values(by='Attrition')
employee_attrition5=employee_attrition4.sort_values(by='Attrition')
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition5[cols]


```

```python
employee_attrition.sort_values(by='MonthlyIncome',ascending=True)
employee_attrition3=employee_attrition.sort_values(by='MonthlyIncome',ascending=True)
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition3[cols]
employee_attrition4=employee_attrition3[employee_attrition3.Education!=0]
employee_attrition4
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition4[cols]
employee_attrition4.sort_values(by='Attrition')
employee_attrition5=employee_attrition4.sort_values(by='Attrition')
cols = ['Attrition','Education','MonthlyIncome']
employee_attrition5[cols]
employee_attrition5.groupby('Education').MonthlyIncome.agg(['count','mean','max','min'])
```
